```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load data into a DataFrame
data = pd.read_csv('Sales_add.csv')

# Create a contingency table of the two features
contingency_table = pd.crosstab(data['Region'], data['Manager'])

# Perform chi-square test of independence
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Determine the significance level (alpha)
alpha = 0.05

# Check if there is a dependency based on the p-value
if p_value < alpha:
    print("There is a dependency between Region and Manager.")
else:
    print("There is no dependency between Region and Manager.")
```
